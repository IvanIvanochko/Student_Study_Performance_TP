## Стовпчаста діаграма

```python
# Bin the math scores into intervals of 10
bins = range(0, 101, 10)
labels = [f"{i}-{i+9}" for i in bins[:-1]]
ds['math_score_bins'] = pd.cut(ds['math_score'], bins=bins, labels=labels, include_lowest=True)

# Adjust the figure size
plt.rcParams["figure.figsize"] = (10, 5)

# Create a count plot using Seaborn with binned math scores
sns.countplot(x="math_score_bins", hue="gender", data=ds, palette="dark")

# Set the title of the plot
plt.title("Count of Math Scores by Gender (Binned by 10)", fontsize=20)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
bins = range(0, 101, 10)
labels = [f"{i}-{i+9}" for i in bins[:-1]]
ds['reading_score_bins'] = pd.cut(ds['reading_score'], bins=bins, labels=labels, include_lowest=True)

# Adjust the figure size
plt.rcParams["figure.figsize"] = (10, 5)

# Create a count plot using Seaborn with binned scores
sns.countplot(x="reading_score_bins", hue="gender", data=ds, palette="dark")

# Set the title of the plot
plt.title("Count of Reading Scores by Gender (Binned by 10)", fontsize=20)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
bins = range(0, 101, 10)
labels = [f"{i}-{i+9}" for i in bins[:-1]]
ds['writing_score_bins'] = pd.cut(ds['writing_score'], bins=bins, labels=labels, include_lowest=True)

# Adjust the figure size
plt.rcParams["figure.figsize"] = (10, 5)

# Create a count plot using Seaborn with binned scores
sns.countplot(x="writing_score_bins", hue="gender", data=ds, palette="dark")

# Set the title of the plot
plt.title("Count of Writing Scores by Gender (Binned by 10)", fontsize=20)

# Show the plot
plt.tight_layout()
plt.show()
```

## Кореляційна матриця

```python
col_for_corr = ['math_score', 'reading_score', 'writing_score']

plt.rcParams["figure.figsize"] = (15,12)
correlation_matrix = ds[col_for_corr].corr()
sns.heatmap(
    correlation_matrix,
    vmax=1,
    square=True,
    annot=True,
    fmt='.2f',
    cmap='GnBu',
    cbar_kws={"shrink": .5},
    robust=True)
plt.title('Correlation Matrix of features', fontsize=8)
plt.show()
```
